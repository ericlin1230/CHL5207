```python
import csv
import gzip
import os
import scipy.io
import scipy.sparse
import pandas as pd
```

```python
matrix_dir = "Wu_etal_2021_BRCA_scRNASeq" # Replace this with the directory of the files
mat = scipy.io.mmread(os.path.join(matrix_dir, "count_matrix_sparse.mtx"))
```

```python
type(mat)
```

```python
# Loads the rows for the datasets
features_path = os.path.join(matrix_dir, "features.tsv")
ifile = open(features_path, 'r')
read = csv.reader(ifile, delimiter="\t")
feature_ids, gene_names, feature_types, barcodes = [], [], [], []
for row in read:
    feature_ids.append(row[0])
    gene_names.append(row[1])
    feature_types.append(row[2])

# Loads the columns for the dataframe
barcodes_path = os.path.join(matrix_dir, "barcodes.tsv")
ifile = open(barcodes_path, 'r')
read = csv.reader(ifile, delimiter="\t")
for row in read:
    barcodes.append(row[0])
```

```python
df = pd.DataFrame.sparse.from_spmatrix(mat)
```

```python
print(df.shape)
print(len(barcodes))
print(len(feature_ids))
```

```python
df.set_axis(gene_names, axis=0, inplace=True)
df.set_axis(barcodes, axis=1, inplace=True)
df.head()
```
